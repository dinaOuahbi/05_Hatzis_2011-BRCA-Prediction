# DataPreprocessing

- minmax
- standard scaler
- robust scaler

```python
X = df.drop('RECIST', axis=1)
y = df['RECIST']
print(X.shape, y.shape)
```

```python
#equilibrer les classes 
#from sklearn.utils import class_weight
#class_weights = class_weight.compute_class_weight('balanced',np.unique(y_train),y_train)
#print(class_weights)
```

```python
#minmax
#no outlier (see distribution)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_sc = scaler.fit_transform(X)
```

```python
print('Average: ',X_sc.mean().round(3))
print('standard deviation', X_sc.std().round(3))
print('maximum: ',X_sc.max().round(3))
print('minimun: ',X_sc.min().round(3))
```

```python
#import models
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, matthews_corrcoef
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
```

```python
#split data
X_sc_train, X_sc_test, y_train, y_test = train_test_split(X_sc, y, random_state=0, test_size=0.25)
```

```python
clf1 = SVC()
clf2 = LogisticRegression(penalty='l2',random_state=0)
clf3 = KNeighborsClassifier(n_neighbors = 5,metric = 'minkowski',p = 2)
clf4 = GaussianNB()
clf5 = DecisionTreeClassifier(criterion = 'entropy',random_state = 0,max_depth=None)
list_model = [clf1,clf2,clf3,clf4,clf5]
```

```python
dict_mcc = {}
dict_acc = {}
dict_score = {}
for model in list_model:
    model.fit(X_sc_train, y_train)
    dict_mcc[model] = matthews_corrcoef(y_test, model.predict(X_sc_test))
    dict_acc[model] = accuracy_score(y_test, model.predict(X_sc_test))
    dict_score[model] = model.score(X_sc_test,y_test)
    print('acc_{} = {}'.format(model, accuracy_score(y_test, model.predict(X_sc_test))))
    print('mcc_{} = {}'.format(model, matthews_corrcoef(y_test, model.predict(X_sc_test))))
    print('score_{} = {}'.format(model, model.score(X_sc_test,y_test)))
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.bar(np.arange(1,6), dict_acc.values(), label='acc', alpha=0.4)
plt.xlabel('models')
plt.ylabel('score')
plt.grid(True)
plt.legend()
plt.subplot(122)
plt.bar(np.arange(1,6), dict_mcc.values(), label='mcc', alpha=0.4)
plt.xlabel('models')
plt.ylabel('score')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#standardization 
from sklearn.preprocessing import StandardScaler
stdscl = StandardScaler()
X_std = stdscl.fit_transform(X)
```

```python
print('Average: ',X_std.mean().round(3))
print('standard deviation', X_std.std().round(3))
print('maximum: ',X_std.max().round(3))
print('minimun: ',X_std.min().round(3))
```

```python
#try models with stdscaler data
X_sc_train, X_sc_test, y_train, y_test = train_test_split(X_std, y, random_state=0, test_size=0.25)

#models in forloop
dict_mcc = {}
dict_acc = {}
for model in list_model:
    model.fit(X_std_train, y_train)
    model.score(X_sc_test,y_test)
    dict_mcc[model] = matthews_corrcoef(y_test, model.predict(X_sc_test))
    dict_acc[model] = accuracy_score(y_test, model.predict(X_sc_test))                                   
    print('acc_{} = {}'.format(model, accuracy_score(y_test, model.predict(X_sc_test))))
    print('mcc_{} = {}'.format(model, matthews_corrcoef(y_test, model.predict(X_sc_test))))
```

# Train test split_original data

```python
#datasets first is shuffle and sparated into 4 batchs
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

# DECISION TREE

## all features and differents depth

```python
#with original data
from sklearn.metrics import matthews_corrcoef, accuracy_score
from sklearn.tree import DecisionTreeClassifier
depth_acc = []
depth_mcc = []
depth = np.arange(2,10)
for d in depth:
    dt_clf = DecisionTreeClassifier(criterion = 'entropy',random_state = 0, max_depth=d)
    dt_clf.fit(X_train, y_train)
    y_pred_dt = dt_clf.predict(X_test)
    acc = accuracy_score(y_test, y_pred_dt)
    depth_acc.append(acc)
    print(' depth {} Acc CART  ===> {} '.format(d, acc))
    mcc = matthews_corrcoef(y_test, y_pred_dt)
    depth_mcc.append(mcc)
    print(' depth {} Mcc CART  ===> {} '.format(d, mcc))
```

```python
plt.figure(figsize=(5,5))

plt.plot(depth, depth_acc, label='acc')
plt.plot(depth, depth_mcc, label='mcc')
plt.legend()
plt.xlabel('depths-dt')
plt.ylabel('scores')
plt.grid(True)
```

## SELECT KBEST

```python
Feature_names = pd.DataFrame(X_sc).columns
list_col = X.columns
```

```python
#freature selection ANOVA
# Feature Selection with Univariate Statistical Tests
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif, chi2

# feature extraction (4 best features)
test = SelectKBest(score_func=f_classif, k=10)
fit = test.fit(X_train, y_train)
# summarize scores
set_printoptions(precision=3)
print(fit.scores_)
best_features = fit.transform(X)
# summarize selected features
best_features.shape
```

```python
test.get_support()
```

```python
#get the index of best features with get_support function ==> boolean mask
index_selected_f = np.array(Feature_names)[test.get_support()]
index_selected_f
```

```python
#get genes name that were selected by f_classif test
for i in index_selected_f:
    print('index {} ===> {}'.format(i, list_col[i]))
```

```python
selector = SelectKBest(chi2, k=10)
best_features_chi2 = selector.fit_transform(X_sc, y)
chi2(X_sc,y)

#get index columns selected
index_selected_chi2 = np.array(Feature_names)[selector.get_support()]
index_selected_chi2
```

```python
#get genes name that were selected by chie2 test
for i in index_selected_chi2:
    print('index {} ===> {}'.format(i, list_col[i]))
```

```python
#check shape of best_feature
print(best_features_chi2.shape)
```

## DT with cross validation

```python
#cross validation : KFOLD
from sklearn.model_selection import cross_val_score, cross_val_predict, StratifiedKFold
dt_clf = DecisionTreeClassifier(criterion = 'entropy',random_state = 0, max_depth=None)
cv = StratifiedKFold(10, random_state=0)
dt_scores = cross_val_score(dt_clf, X, y, cv=cv)
ypred_dt = cross_val_predict(dt_clf, X, y, cv=cv)
print(dt_scores)
print(np.mean(dt_scores))
print(matthews_corrcoef(y, ypred_dt))
```

# dt with best features

```python
#cross validation : KFOLD with best_feature
from sklearn.model_selection import cross_val_score, cross_val_predict, StratifiedKFold
dt_clf = DecisionTreeClassifier(criterion = 'entropy',random_state = 0, max_depth=None)
cv = StratifiedKFold(10, random_state=0)
dt_scores = cross_val_score(dt_clf, best_features, y, cv=cv)
ypred_dt = cross_val_predict(dt_clf, best_features, y, cv=cv)
print(dt_scores)
print(np.mean(dt_scores))
print(matthews_corrcoef(y, ypred_dt))
```

# STACK

```python
#combine models in just one (stack model)
from sklearn.ensemble import StackingClassifier
#list of models
estimator_list = [
    ('knn',knn_clf),
    ('svm_rbf',svc_clf),
    ('rf',rf_clf),
    ('dt',dt_clf)]
stack_model = StackingClassifier(estimators=estimator_list,
                                 final_estimator=lr_clf)

stack_model.fit(X_train, y_train)
ypred_sm = stack_model.predict(X_test)
print('Acc stack model ===> ',accuracy_score(y_test, ypred_sm))
print('Mcc stack model ===> ',matthews_corrcoef(y_test, ypred_sm))
```
