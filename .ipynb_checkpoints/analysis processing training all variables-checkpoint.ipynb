# SPLIT

```python
from sklearn.model_selection import train_test_split
X = data.drop('RECIST', axis=1)
y = data['RECIST']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

# BUILT MODELS

```python
from sklearn import metrics
#KNN model
from sklearn.metrics import accuracy_score 
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier(n_neighbors =5,
                               metric = 'minkowski',
                               p = 2)
knn_clf.fit(X_train, y_train)
print('knn-score: ',knn_clf.score(X_test, y_test))
y_pred_knn = knn_clf.predict(X_test)
print('ACC : ',accuracy_score(y_test, y_pred_knn))
print('mcc:', metrics.matthews_corrcoef(y_test, y_pred_knn))
```

```python
#trouver le meilleur hyperparams n-neighbord avec une boucle
train_acc = []
test_acc = []
neighbors_setting = np.arange(1,30)
for n in neighbors_setting:
    knn = KNeighborsClassifier(n_neighbors=n)
    knn.fit(X_train,y_train)
    train_acc.append(knn.score(X_train, y_train))
    test_acc.append(knn.score(X_test, y_test))
```

```python
#learning curve
plt.subplot(121)
plt.plot(train_acc, c='red')
plt.title('train_acc')
plt.subplot(122)
plt.plot(test_acc, c='green')
plt.title('test_acc')
```

```python
X.reset_index(inplace=True)
```

```python
X.drop('patient_ID', axis=1, inplace=True)
```

```python
y = pd.DataFrame(data=y)
y.reset_index(inplace=True)
y.drop('patient_ID', axis=1, inplace=True)
```

```python
def get_score(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

# KNN

```python
#cross validation on KNN model
from sklearn.metrics import matthews_corrcoef
from sklearn.model_selection import StratifiedKFold
knn_scores = []
mcc_scores = []
skf = StratifiedKFold(n_splits=10, random_state=0)
skf.get_n_splits(X, y)

print(skf)

for train_index, test_index in skf.split(X, y):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X.iloc[train_index, :], X.iloc[test_index, :]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    #print(X_train.shape)
    knn_clf.fit(X_train, y_train)
    y_pred = knn_clf.predict(X_test)
    sc = matthews_corrcoef(y_test, y_pred)
    mcc_scores.append(sc)
    knn_scores.append(get_score(KNeighborsClassifier(n_neighbors=5), X_train, X_test, y_train, y_test))
print('moyenne mcc =====>',np.mean(mcc_scores))
print('moyenne score-knn =====>',np.mean(knn_scores))
```

```python
print('knn scores (10 scores for each fold):\n',knn_scores)
print('mcc scores (10 scores for each fold):\n',mcc_scores)
print('Moyenne des scores knn ===> ',np.mean(knn_scores))
print('Moyenne des scores mcc ===> ',np.mean(mcc_scores))
```

```python
#list of available scoring
from sklearn.metrics import SCORERS
SCORERS.keys()
```

```python
#StratofiedKFold / KNN / mcc
from sklearn.model_selection import cross_val_score, cross_val_predict
cv = StratifiedKFold(10, random_state=0)
scoreKnn = cross_val_score(knn_clf, X, y, cv=cv)
ypred_knn_cv = cross_val_predict(knn_clf, X, y, cv=cv)
print(scoreKnn)
print(np.average(scoreKnn))
print('MCC-score ==> ',matthews_corrcoef(y, ypred_knn_cv))
```

# Logistic regression

```python
#LOGISTIC REG
from sklearn.linear_model import LogisticRegression

lr_scores = []
lr_mcc = []

lr_clf = LogisticRegression(penalty='l2',random_state=0)
lr_clf.fit(X_train, y_train)

skf = StratifiedKFold(n_splits=10, random_state=0)
skf.get_n_splits(X, y)

print(skf)

for train_index, test_index in skf.split(X, y):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X.iloc[train_index, :], X.iloc[test_index, :]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    #print(X_train.shape)
    lr_clf.fit(X_train, y_train)
    y_pred = lr_clf.predict(X_test)
    sc = matthews_corrcoef(y_test, y_pred)
    lr_mcc.append(sc)
    lr_scores.append(get_score(lr_clf, X_train, X_test, y_train, y_test))
print('moyenne mcc =====>',np.mean(lr_mcc))
print('moyenne score-lr =====>',np.mean(lr_scores))
```

# ROC and AUC

```python
from sklearn.metrics import roc_curve, roc_auc_score

lr_probs = lr_clf.predict_proba(X_test)
#svc_probs = svc_clf.predict_proba(X_test)
rf_probs = rf_clf.predict_proba(X_test)
knn_probs = knn_clf.predict_proba(X_test)

# generate a no skill prediction (majority class)
ns_probs = [0 for _ in range(len(y_test))]

# keep probabilities for the positive outcome only
lr_probs = lr_probs[:, 1]
#svc_probs = svc_probs[:, 1]
rf_probs = rf_probs[:, 1]
knn_probs = knn_probs[:, 1]

# calculate scores
ns_auc = roc_auc_score(y_test, ns_probs)
lr_auc = roc_auc_score(y_test, lr_probs)
#svc_auc = roc_auc_score(y_test, svc_probs)
rf_auc = roc_auc_score(y_test, rf_probs)
knn_auc = roc_auc_score(y_test, knn_probs)

# summarize scores
print('No Skill: ROC AUC=%.3f' % (ns_auc))
print('Logistic: ROC AUC=%.3f' % (lr_auc))
#print('SVC: ROC AUC=%.3f' % (svc_auc))
print('RF: ROC AUC=%.3f' % (rf_auc))
print('knn: ROC AUC=%.3f' % (knn_auc))

# calculate roc curves
ns_fpr, ns_tpr, _ = roc_curve(y_test, ns_probs)
lr_fpr, lr_tpr, _ = roc_curve(y_test, lr_probs)
#svc_fpr, svc_tpr, _ = roc_curve(y_test, svc_probs)
rf_fpr, rf_tpr, _ = roc_curve(y_test, rf_probs)
knn_fpr, knn_tpr, _ = roc_curve(y_test, knn_probs)

# plot the roc curve for the model
plt.plot(ns_fpr, ns_tpr, linestyle='--', label='No Skill')
plt.plot(lr_fpr, lr_tpr, marker='.', label='Logistic')

#plt.plot(svc_fpr, svc_tpr, linestyle='--', label='svc')
plt.plot(rf_fpr, rf_tpr, marker='.', label='rf')
plt.plot(knn_fpr, knn_tpr, marker='.', label='knn')

# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

# show the legend
plt.legend()
# show the plot
plt.show()
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
depth_acc = []
depth_mcc = []
depth = np.arange(2,10)
for d in depth:
    dt_clf = DecisionTreeClassifier(criterion = 'entropy',random_state = 0, max_depth=d)
    dt_clf.fit(X_train, y_train)
    y_pred_dt = dt_clf.predict(X_test)
    acc = accuracy_score(y_test, y_pred_dt)
    depth_acc.append(acc)
    print(' depth {} Acc CART  ===> {} '.format(d, acc))
    mcc = matthews_corrcoef(y_test, y_pred_dt)
    depth_mcc.append(mcc)
    print(' depth {} Mcc CART  ===> {} '.format(d, mcc))
```

```python
plt.figure(figsize=(7,7))
plt.plot(depth, depth_acc, label='acc')
plt.plot(depth, depth_mcc, label='mcc')
plt.xlabel('depth')
plt.ylabel('Score')
plt.legend()
plt.show()
```
